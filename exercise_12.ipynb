2. Consider the following function:

y(x) = $\sqrt{\frac{100(1 - 0.01x^2)^2 + 0.02x^2}{(1 - x^2)^2 + 0.1x^2}}$

Generate a 2√ó2
 subplot of ùë¶(ùë•)
 for 0‚â§ùë•‚â§100
 using plot, semilogx, semilogy, and loglog. Use a fine enough discretization in x to make the plot appear smooth. Give each plot axis labels and a title. Turn the grid on. Which plot seems to convey the most information?.

```python
x = np.linspace(0,100, 200)
num_left = 100 * ( 1 - 0.01*(x**2) )**2
num_right = 0.02*(x**2)
num = num_left + num_right

den_left = (1 - x**2)**2
den_right = 0.1*(x**2)
den = den_left + den_right

y = np.sqrt(num/den)

fig, axes = plt.subplots(2, 2)
axes[0,0].plot(x, y)
axes[0,0].set_title('y by x (both on normal scale)')
axes[0,0].set_xlabel('x')
axes[0,0].set_ylabel('y')

axes[0,1].semilogx(x, y)
axes[0,1].set_title('y by x (x on log scale)')
axes[0,1].set_xlabel('x')
axes[0,1].set_ylabel('y')


axes[1,0].semilogy(x, y)
axes[1,0].set_title('y by x (y on log scale)')
axes[1,0].set_xlabel('x')
axes[1,0].set_ylabel('y')


axes[1,1].loglog(x, y)
axes[1,1].set_title('y by x (both on log scale)')
axes[1,1].set_xlabel('x')
axes[1,1].set_ylabel('y')


plt.tight_layout()
plt.show()
```

7. Write a function my_polygon(n) that plots a regular polygon with n sides and radius 1. Recall that the radius of a regular polygon is the distance from its centroid to the vertices. Use plt.axis(‚Äòequal‚Äô) to make the polygon look regular. Remember to give the axes a label and a title. You can use plt.title to title the plot according to the number of sides. Hint: This problem is significantly easier if you think in polar coordinates. Recall that a complete revolution around the unit circle is 2ùúã
 radians. Note: The first and last point on the polygon should be the point associated with the polar coordinate angles, 0
 and 2ùúã
, respectively.

There are two steps that are needed to compute the coordinates of a polygon:

First assuming the center is [0,0] we compute each coordinate using the formula:

[r * cos($\frac{2œÄi}{n}$), r * sin($\frac{2œÄi}{n}$)]

where __r__ is radius and __n__ are the number of sides. $\frac{2œÄ}{n}$ is the angle between two corodinates.

if center was not [0,0] then the formula is:

[x + r * cos($\frac{2œÄi}{n}$), y + r * sin($\frac{2œÄi}{n}$)]

where x,y would be the coordinates of the center. Here, both are 0,0 i.e. the origin.

The next step is to rotate them $90^o$ counterclockwise to ensure the polygon is symmetrical along the y-axis.

Thus, we add $\frac{œÄ}/{2}$ to each coordinate to rotate them.

source: https://stackoverflow.com/questions/36023612/calculate-coordinates-of-a-polygon-using-radius-and-centerpoint

```python
def polygon(n, r = 1):

  coord = []

  for i in range(1,n+1):
    # coord.append([r * np.cos((2 * np.pi * i)/n), r * np.sin((2 * np.pi * i)/n)])
    angle = np.pi/2 + (2 * np.pi * i)/n # we will add the rotation pi/2 to the angle (2 * pi * i)/n to make computation easier to read
                                        # this is same as the reference code source provided above in C
    coord.append([r * np.cos(angle), r * np.sin(angle)])

  # plot each coordinate points
  for pt in coord:
    plt.plot(pt[0], pt[1], 'bo')

  # plot lines connecting the coordinate [points]
  for iter in range(n-1):
    plt.plot([coord[iter][0], coord[iter+1][0]], [coord[iter][1], coord[iter+1][1]], 'b-')

  # we plotted n - 1 coordinates above. Now we join the first and the last coordinates
  plt.plot([coord[0][0], coord[n-1][0]], [coord[0][1], coord[n-1][1]], 'b-')
  plt.axis('equal')
  plt.show()
```

```python
# additional case with a 4 sided polygon, which should be diamond shaped
polygon(4)
```

```python
# 6 sided polygon
polygon(6)
```

10. Assume you have three points at the corner of an equilateral triangle,
P1 = (0,0), P2 = (0.5, $\frac{\sqrt{2}}{2}$), P3 = (1,0). Now you want to generate another set of points ùëùùëñ=(ùë•ùëñ,ùë¶ùëñ) such that ùëù1=(0,0) and ùëùùëñ+1 is the midpoint between ùëùùëñ and ùëÉ1 with 33% probability, the midpoint between ùëùùëñ and ùëÉ2 with 33% probability, and the midpoin between ùëùùëñ and ùëÉ3 with 33% probability. Write a function
my_sierpinski(n) that generates the points ùëùùëñ for ùëñ=1,‚ãØ,ùëõ. The function should make a plot of the points using blue dots (i.e., ‚Äòb.‚Äô as the third argument to the plt.plot function).


The solution to this problem is simple:

We first generate a random number between 0 and 1 drawn from a uniform distribution. If the number is between 0 and 0.3333, we compute and plot the midpoint of P1 and pi, if the value is between 0.3334 and 0.6666 then we compute and plot the midpoint between P2 and pi, and in all other scenarios which number greater than 0.6666 we compute and plot the midpoint between P3 and pi.

Initial pi is [0,0].

```python
def my_sierpinski(n):
  p1 = [0,0]
  p2 = [0.5, np.math.sqrt(2)/2]
  p3 = [1,0]

  arr = np.zeros(2)

  for iter in range(n):
    rand_num = np.random.rand(1)[0]

    if rand_num > 0 and rand_num <= 0.3333:
      arr[0] = (arr[0] + p1[0])/2
      arr[1] = (arr[1] + p1[1])/2

    elif rand_num >= 0.3334 and rand_num <= 0.6666:
      arr[0] = (arr[0] + p2[0])/2
      arr[1] = (arr[1] + p2[1])/2

    else:
      arr[0] = (arr[0] + p3[0])/2
      arr[1] = (arr[1] + p3[1])/2

    plt.plot(arr[0], arr[1], 'bo', markersize = 1.5)

  plt.show()
```

```python
# test case 1: n = 100
my_sierpinski(100)
```

```python
# test case 2: n = 10000
my_sierpinski(10000)
```

11. Assume you are generating a set of points (ùë•ùëñ,ùë¶ùëñ)
 where ùë•1=0
 and ùë¶1=0
. The points (ùë•ùëñ,ùë¶ùëñ)
 for ùëñ=2,‚ãØ,ùëõ
 is generated according to the following probabilistic relationship:

With 1% probability:

xi=0

yi = $0.16y_{i-1}$

With 7% probability:

xi = $0.2x_{i-1} - 0.26y_{i-1}$

yi = $0.23x_{i-1} + 0.22y_{i-1} + 1.6$

With 7% probability:

xi = $-0.15x_{i-1} + 0.28y_{i-1}$

yi = $0.26x_{i-1} + 0.24y_{i-1} + 0.44$

With 85% probability:

xi = $0.85x_{i-1} + 0.04y_{i-1}$

yi = $-0.04x_{i-1} + 0.85y_{i-1} + 1.6$


Write a function my_fern(n) that generates the points (xi,yi) for i=1,‚Ä¶,n
and plots them using blue dots. Also use plt.axis('equal') and plt.axis('off') to make the plot looks nicer.

```python
def my_fern(n):

  pts = [(0,0)]

  for iter in range(n+1):
    rand_num = np.random.rand(1)[0]
    x_prev = pts[-1][0]
    y_prev = pts[-1][1]

    if rand_num <= 0.01:
      xi = 0
      yi = 0.16*y_prev

    elif rand_num <= 0.08:
      xi = 0.2*x_prev - 0.26*y_prev
      yi = 0.23*x_prev + 0.22*y_prev + 1.6

    elif rand_num <= 0.15:
      xi = -0.15*x_prev + 0.28*y_prev
      yi = 0.26*x_prev + 0.24*y_prev + 0.44

    else:
      xi = 0.85*x_prev + 0.04*y_prev
      yi = -0.04*x_prev + 0.85*y_prev + 1.6

    pts.append((xi,yi))

    plt.plot([xi], [yi], 'bo', markersize = 1, alpha = 0.5)
    plt.axis('equal')
    plt.axis('off')

  plt.title(f'My fern with {n} iterations')
  plt.show()
```

```python
my_fern(100)
```

12. Write a function my_parametric_plotter (x,y,t) where x and y are function objects x (t) and y (t), respectively, and t is a one-dimensional~array. The function my_parametric_plotter should produce the curve (x(t), y(t), t) in a three-dimensional plot. Be sure to give your plot a title and axis labels.

```python
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D


def my_parametric_plotter(x, y, t):
  fig = plt.figure(figsize = (12,6))

  ax = fig.add_subplot(1,1,1, projection='3d')
  ax.plot(x(t), y(t), t)
  ax.grid(visible = False)
  ax.set_title('My parametric plot')
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_zlabel('t', labelpad = 2)
  plt.show()
```

13. Write a function my_surface_plotter(f, x, y, option) where f is a function object f (x,y). The function my_surface_plotter should produce a 3D surface plot of f (x,y) using plot_surface if option is the string ‚Äòsurface‚Äô. It should produce a contour plot of f (x,y) if the option is the string ‚Äòcontour‚Äô. Assume that x and y are one-dimensional arrays or lists. Remember to give the plot a title and axis labels.

```python
from matplotlib import cm
```

```python
def my_surface_plotter(f, x, y, option):
  X, Y = np.meshgrid(x,y)
  Z = f(X,Y)

  if option == 'surface':
    fig = plt.figure(figsize = (12,6))
    ax = fig.add_subplot(1,1,1, projection = '3d')
    ax.plot_surface(X, Y, Z, cmap = cm.coolwarm)
    ax.set_title('My surface Plot')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z', labelpad = -1)

  elif option == 'contour':
    plt.contour(X, Y, Z)
    plt.title('My contour Plot')
    plt.xlabel('x')
    plt.ylabel('y')

  plt.show()
```

```python
f = lambda x, y: np.cos(y)*np.sin(np.exp(x))
my_surface_plotter(f, np.linspace(-1, 1, 20), np.linspace(-2, 2, 40), 'surface')
my_surface_plotter(f, np.linspace(-1, 1, 20), np.linspace(-2, 2, 40), 'contour')
```
