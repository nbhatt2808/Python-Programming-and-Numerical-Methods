10. A year is considered to be 365 days long. However, a more exact figure is 365.24 days. As a consequence, if we held to the standard 365-day year, we would gradually lose that fraction of the day over time, and seasons and other astronomical events would not occur as expected. A leap year is a year that has an extra day, February 29, to keep the timescale on track. Leap years occur on years that are exactly divisible by 4, unless it is exactly divisible by 100, unless it is divisible by 400. For example, the year 2004 is a leap year, the year 1900 is not a leap year, and the year 2000 is a leap year. Compute the number of leap years between the years 1500 and 2010.


Here's how the sequence works.

If a year is not divisible by 4 then it's not a leap year.

If a year is divisible by 4 then: <br>
- If the year is not divisible by 100, then its a leap year
- else if the year is divisible by 100, then its a leap year only if its also divisible by 400.

```python
start_year = 1500
end_year = 2010
count_leaps = 0

# loop over years and move by 4 as 1500 is divisible by 4 and to be a leap year each year has to be divisible by 4
for year in range(start_year, end_year + 1, 4):
  
  # check if divisible by 4
  if year % 4 == 0:
    
    # if the year is not divisible by 100 then its a leap year
    if year % 100 != 0:
      print(f'{year} is divisble by 4 but not by 100 and thus is a leap year.')
      count_leaps += 1
    
    # if the year is divisible by 100 then it must also be divisible by 400 to be a leap year
    else:
      if year % 400 == 0:
        print(f'{year} is divisble by 4, and by 100 and also by 400 and thus is a leap year.')
        count_leaps += 1

print()
print(f'Total leap years between {start_year} and {end_year}: {count_leaps}')
```

18. Let P and Q be logical expressions. De Morgan‚Äôs rule states that 

ùëÅùëÇùëá (ùëÉ ùëÇùëÖ ùëÑ) = (ùëÅùëÇùëá ùëÉ) ùê¥ùëÅùê∑ (ùëÅùëÇùëá ùëÑ) and <br>
ùëÅùëÇùëá (ùëÉ ùê¥ùëÅùê∑ ùëÑ) = (ùëÅùëÇùëá ùëÉ) ùëÇùëÖ (ùëÅùëÇùëá ùëÑ). 

Generate the truth tables for each statement to show that De Morgan‚Äôs rule is always true.

```python
P = 2 == 1+1 # true statement
Q = 4+2 == 7 # false statement

# verifying de morgan's rule
print(not (P or Q) == (not P) & (not Q))
print(not (P & Q) == (not P) or (not Q))
```

```python
# creating truth tables for P and Q
import numpy as np
import pandas as pd

arr = np.array([[P & Q, P & (not Q)], [not P & Q, not P & (not Q)]], dtype = int)
truth_table = pd.DataFrame(arr)

truth_table.index = pd.MultiIndex.from_tuples([('P', int(P)), ('P', int(not P))])
truth_table.columns = pd.MultiIndex.from_tuples([('Q', int(not Q)), ('Q', int(Q))])

truth_table
```

20. Construct an equivalent logical expression for OR using only AND and NOT.

```python
arr = np.arange(10)
arr
```

```python
# extracting numbers divisible by either 3 or 4
np.where((arr % 3 == 0) | (arr % 4 == 0))
```

From the quora post it's mentioned that NAND gate is equivalent to OR when using inverse inputs.

NAND stands for negative AND and returns 0 when both inputs are 1. (geeks for geeks blog referenced above). 

```python
# code taken from: https://www.geeksforgeeks.org/logic-gates-in-python/
# only when one of the inputs are false we return 1
def NAND (a, b):
    if a == 1 and b == 1:
        return False
    else:
        return True
```

```python
# replicating OR using NAND. Previously we used or filter values dividible by 3 or 4.
# Here, inverse inputs would mean value not divisible by 3 and 4, which means val % 3 != 0 and val % 4 != 0
logic = [NAND(val % 3 != 0, val % 4 != 0) for val in arr]
arr[np.where(np.array(logic) == 1)[0]]
```

21. Construct an equivalent logical expression for AND using only OR and NOT.

Similarly, equivalent of AND is NOR or negative OR when using inverse inputs. (quora post referenced above)

```python
arr = np.arange(10)
arr
```

```python
# filtering values between 2 and 7 (inclusive)
np.where((arr >= 2) & (arr <= 7))
```

```python
# code for NOR taken from: https://www.geeksforgeeks.org/logic-gates-in-python/
# only when both inputs are false we return the value as 1 else all are 0
def NOR(a, b):
    if(a == 0) and (b == 0):
        return 1
    elif(a == 0) and (b == 1):
        return 0
    elif(a == 1) and (b == 0):
        return 0
    elif(a == 1) and (b == 1):
        return 0
```

```python
# providing inverted inputs for NAND, which is val < 2 and val > 7. only where both are false we return True
logic = [NOR(val < 2, val > 7) for val in arr]
arr[np.where(np.array(logic) == 1)[0]]
```
